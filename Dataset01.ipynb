```python
#Média das Quantidade e Valores apresentad, aprovada e apoiada dos dados. 
mediaQApresentada = df['Quantidade Apresentada'].mean() # Média da coluna
print("Média da Quantidade Apresentada:", mediaQApresentada)


mediaQAprovada = df['Quantidade Aprovada'].mean() # Média da coluna
print("Média da Quantidade Aprovada:", mediaQAprovada)

mediaQApoiada = df['Quantidade Apoiada'].mean() # Média da coluna
print("Média da Quantidade Apoiada:", mediaQApoiada)

mediaVApresentado = df['Vl.Apresentado'].mean() # Média da coluna
print("Média do Valor Apresentado:", mediaVApresentado)

mediaVAprovado = df['Vl. Aprovado'].mean() # Média da coluna
print("Média do Valor Aprovado:", mediaVAprovado)

mediaVApoiado = df['Vl. Apoiado'].mean() # Média da coluna
print("Média do Valor Apoiado:", mediaVApoiado)
```

```python
#Mediana das Quantidade e Valores apresentad, aprovada e apoiada dos dados. 
medianaQApresentada = df['Quantidade Apresentada'].median() # Mediana da coluna
print("Mediana da Quantidade Apresentada:", medianaQApresentada)


medianaQAprovada = df['Quantidade Aprovada'].median() # Mediana da coluna
print("Mediana da Quantidade Aprovada:", medianaQAprovada)

medianaQApoiada = df['Quantidade Apoiada'].median() # Mediana da coluna
print("Mediana da Quantidade Apoiada:", medianaQApoiada)

medianaVApresentado = df['Vl.Apresentado'].median() # Mediana da coluna
print("Mediana do Valor Apresentado:", medianaVApresentado)

medianaVAprovado = df['Vl. Aprovado'].median() # Mediana da coluna
print("Mediana do Valor Aprovado:", medianaVAprovado)

medianaVApoiado = df['Vl. Apoiado'].median() # Mediana da coluna
print("Mediana do Valor Apoiado:", medianaVApoiado)
```

```python
#Moda das Quantidade e Valores apresentad, aprovada e apoiada dos dados. 
modaQApresentada = df['Quantidade Apresentada'].mode() # Moda da coluna
print("Moda da Quantidade Apresentada:", modaQApresentada)


modaQAprovada = df['Quantidade Aprovada'].mode() # Moda da coluna
print("Moda da Quantidade Aprovada:", modaQAprovada)

modaQApoiada = df['Quantidade Apoiada'].mode() # Moda da coluna
print("Moda da Quantidade Apoiada:", modaQApoiada)

modaVApresentado = df['Vl.Apresentado'].mode() # Moda da coluna
print("Moda do Valor Apresentado:", modaVApresentado)

modaVAprovado = df['Vl. Aprovado'].mode() # Moda da coluna
print("Moda do Valor Aprovado:", modaVAprovado)

modaVApoiado = df['Vl. Apoiado'].mode() # Moda da coluna
print("Moda do Valor Apoiado:", modaVApoiado)
```

```python
#Variança das Quantidade e Valores apresentad, aprovada e apoiada dos dados. 
varianca = df['Quantidade Apresentada'].var()
print("Variânça Quantidade Apresentada:", varianca)

varianca = df['Quantidade Aprovada'].var()
print("Variânça Quantidade Aprovada:", varianca)

varianca = df['Quantidade Apoiada'].var()
print("Variânça Quantidade Apoiada:", varianca)

varianca = df['Vl.Apresentado'].var()
print("Variânça Valor Apresentado:", varianca)

varianca = df['Vl. Aprovado'].var()
print("Variânça Valor Aprovado:", varianca)

varianca = df['Vl. Apoiado'].var()
print("Variânça Valor Apoiado:", varianca)
```

```python
#Desvio padrão das Quantidade e Valores apresentada, aprovada e apoiada dos dados. 
desvio = df['Quantidade Apresentada'].std()
print("Devio padrão Quantidade Apresentada:", desvio)

desvio = df['Quantidade Aprovada'].std()
print("Desvio padrão Quantidade Aprovada:", desvio)

desvio = df['Quantidade Apoiada'].std()
print("Desvio padrão Quantidade Apoiada:", desvio)

desvio = df['Vl.Apresentado'].std()
print("Desvio padrão Valor Apresentado:", desvio)

desvio = df['Vl. Aprovado'].std()
print("Desvio padrão Valor Aprovado:", desvio)

desvio = df['Vl. Apoiado'].std()
print("Desvio padrão Valor Apoiado:", desvio)
```

```python
#Quartis das Quantidade e Valores apresentada, aprovada e apoiada dos dados. 
Q1qa = df['Quantidade Apresentada'].quantile(0.25)
Q2qa = df['Quantidade Apresentada'].quantile(0.5)
Q3qa = df['Quantidade Apresentada'].quantile(0.75)
print('Primeiro quartil Quantidade Apresentada:', Q1qa)
print('Segundo quartil Quantidade Apresentada (Mediana):', Q2qa)
print('Terceiro quartil Quantidade Apresentada:', Q3qa)
#################################################
Q1ap = df['Quantidade Aprovada'].quantile(0.25)
Q2ap = df['Quantidade Aprovada'].quantile(0.5)
Q3ap = df['Quantidade Aprovada'].quantile(0.75)
print('Primeiro quartil Quantidade Aprovada:', Q1ap)
print('Segundo quartil Quantidade Aprovada (Mediana):', Q2ap)
print('Terceiro quartil Quantidade Aprovada:', Q3ap)
###########################################################
Q1apo = df['Quantidade Apoiada'].quantile(0.25)
Q2apo = df['Quantidade Apoiada'].quantile(0.5)
Q3apo = df['Quantidade Apoiada'].quantile(0.75)
print('Primeiro quartil Quantidade Apoiada:', Q1apo)
print('Segundo quartil Quantidade Apoiada (Mediana):', Q2apo)
print('Terceiro quartil Quantidade Apoiada:', Q3apo)
##########################################################
Q1va = df['Vl.Apresentado'].quantile(0.25)
Q2va = df['Vl.Apresentado'].quantile(0.5)
Q3va = df['Vl.Apresentado'].quantile(0.75)
print('Primeiro quartil Valor apresentado:', Q1va)
print('Segundo quartil Valor apresentado (Mediana):', Q2va)
print('Terceiro quartil Valor apresentado:', Q3va)
########################################################
Q1vap = df['Vl. Aprovado'].quantile(0.25)
Q2vap = df['Vl. Aprovado'].quantile(0.5)
Q3vap = df['Vl. Aprovado'].quantile(0.75)
print('Primeiro quartil Valor aprovado:', Q1vap)
print('Segundo quartil Valor aprovado (Mediana):', Q2vap)
print('Terceiro quartil Valor aprovado:', Q3vap)
#############################################
Q1vapo = df['Vl. Apoiado'].quantile(0.25)
Q2vapo = df['Vl. Apoiado'].quantile(0.5)
Q3vapo = df['Vl. Apoiado'].quantile(0.75)
print('Primeiro quartil Valor apoiado:', Q1vapo)
print('Segundo quartil Valor apoiado (Mediana):', Q2vapo)
print('Terceiro quartil Valor apoiado:', Q3vapo)


```
